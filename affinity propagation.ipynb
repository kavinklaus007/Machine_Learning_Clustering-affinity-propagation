```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
import scipy.cluster.hierarchy as sch
```

```python
import sklearn
from sklearn.cluster import AffinityPropagation
clustering = AffinityPropagation(random_state=5)
re=clustering.fit(X)
labels = clustering.labels_

plt.figure(figsize=(10, 7))
plt.scatter( X[:, 0],X[:, 1], c=labels, cmap='rainbow', s=50)
plt.scatter(clustering.cluster_centers_[:, 0], clustering.cluster_centers_[:, 1], 
            color='black', marker='x', s=500, label='Centers')
plt.title('Affinity Propagation Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 5)
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised['Cluster_group']=y_kmeans
```
